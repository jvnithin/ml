```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/sonar.all-data.csv',header=None)
df.head()
```

```python
df[60].value_counts()
```

```python
df.isnull().sum()
```

```python
df.groupby(60).mean()
```

```python
x=df.drop(columns=60,axis=1)
y=df[60]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1,stratify=y)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_train)
score= accuracy_score(predict,y_train)
print(score)
test_predict=model.predict(x_test)
score2=accuracy_score(test_predict,y_test)
print(score2)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv(r'/content/diabetes.csv')
```

```python
df.head()
```

```python
stand=StandardScaler()
# df.std()
stand.fit_transform(df)
df.std()
```

```python
df.describe()
```

```python
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
scaler=StandardScaler()
# print(df.std())
scaler.fit(x)
standardized=scaler.transform(x)
print(standardized.ndim)
```

```python
x=standardized
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=2)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
train_data=classifier.predict(x_train)
res=accuracy_score(train_data,y_train)
print(res)
test_data=classifier.predict(x_test)
res1=accuracy_score(test_data,y_test)
print(res1)
```
